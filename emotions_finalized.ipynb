```python
data['overall'] = responses.map(analyse_tone)
```

```python
data['overall'] 
```

```python
overall_measures= []

for item in data['overall']:
  #print(item)
  all_tones=[]
  if not item['tones']:
    tone_type='None'
    em_quot= 'None'
    all_tones= [tone_type,em_quot]

  else:
    for i in range(len(item['tones'])):

      tone_type=item['tones'][i]['tone_name']
      #all_tones.append(tone_type)

      em_quot= str(round(   ((item['tones'][i]['score']) *100),2)   )   + '%'
      combo= [tone_type,em_quot]
      all_tones.append(combo)
    #em_quot= str((round(item['tones'][0]['score'])*100),2)+'%'
  overall_measures.append(all_tones)
  #emotional_quotient.append(em_quot)
```
