```python
import spacy
from spacy.matcher import PhraseMatcher
from spacy_langdetect import LanguageDetector
import numpy as np
import pandas as pd
```

```python
nlp = spacy.load("en_core_web_md")

## Pattern matching for gender
matcher = PhraseMatcher(nlp.vocab)
```

```python
data = pd.read_csv('with_emotional_quot.csv')
```

```python
data.drop('overall',axis=1)
```

```python
female_list = ["He","he","him","Him","obgyn","His","his","husband","Husband","boy","Boy","Boyfriend","Boy friend","boyfriend","bf","herpes","break out","female","my vagina","vagina","vaginal"]
male_list = ["She","her","Her","wife","Wife","girl","Girl","Girlfriend","Girl friend","girlfriend","gf","Gf","gay","my penis","male"]
```

```python
male_patterns=[nlp(text) for text in male_list]
matcher.add('gender',None,*male_patterns)
```

```python
data['Gen'] = data['Description'].str.lower().apply(lambda x:matcher(nlp(x)))

data['Gender'] = np.where(data.Gen.str.len() > 0,'Male','')
```

```python
data.head()
```

```python
data = data.drop(['Gen'],axis = 1)
```

```python
matcher1 = PhraseMatcher(nlp.vocab)

female_patterns = [nlp(text) for text in female_list]
matcher1.add("female",None,*female_patterns)

data['Gen'] = data['Description'].str.lower().apply(lambda x:matcher1(nlp(x)))

data['Gender'] = np.where(data.Gen.str.len() > 0, 'Female', data['Gender'])
```

```python
final_data = data.drop('Gen',axis=1)
final_data.to_csv('with_emotion_quotient_final.csv')
final_data['Gender'].value_counts()
```
